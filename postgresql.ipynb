```python
import pandas as pd
from sqlalchemy import create_engine #import library database
```

```python
# Bentuk pertama

import psycopg2

engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost/northwind")

conn = psycopg2.connect("dbname = northwind user = postgres password = 'postgres'")

cursor = conn.cursor()
    # Print PostgreSQL Connection properties
print( conn.get_dsn_parameters(),"\n")

# Print PostgreSQL version
cursor.execute("SELECT version();")
record = cursor.fetchone()
print("You are connected to -", record, "\n")
```

```python
df = """select * from products"""
df = pd.read_sql_query(df, conn)
df.head()
```

```python
# Bentuk kedua
# username:passwords@localhost/databases
connection = create_engine("postgresql://postgres:postgres@localhost/northwind")
```

Kode pertama menggunakan modul psycopg2 untuk membuat koneksi ke database PostgreSQL dan membuka cursor untuk mengeksekusi perintah SQL. Sedangkan pada kode kedua menggunakan modul sqlalchemy dan metode create_engine untuk membuat koneksi ke database PostgreSQL.

Kedua modul tersebut berbeda dalam cara mereka berinteraksi dengan database. psycopg2 adalah modul yang menyediakan antarmuka Python untuk PostgreSQL, sementara sqlalchemy adalah modul ORM (Object-Relational Mapping) yang menyediakan cara yang lebih abstrak untuk berinteraksi dengan database. sqlalchemy memungkinkan pengguna untuk mengakses database melalui objek Python, membuat koneksi dan mengeksekusi perintah SQL, sedangkan psycopg2 memungkinkan pengguna untuk melakukan koneksi dan operasi basis data pada PostgreSQL secara langsung dengan bahasa Python.

Dalam hal ini, kode pertama menggunakan modul psycopg2 dan mengatur koneksi ke database PostgreSQL menggunakan fungsi psycopg2.connect(), sedangkan kode kedua menggunakan modul sqlalchemy dan metode create_engine() untuk membuat koneksi ke database PostgreSQL. Keduanya memiliki fungsi yang sama yaitu untuk membuat koneksi ke database PostgreSQL, tetapi cara penggunaannya berbeda karena perbedaan dalam modul yang digunakan.

Selain itu, sqlalchemy juga menyediakan fitur-fitur ORM dan integrasi dengan berbagai jenis database lainnya, sehingga memungkinkan pengguna untuk menggunakan Python sebagai antarmuka umum untuk berinteraksi dengan berbagai jenis database. Sedangkan psycopg2 lebih fokus pada koneksi dan operasi basis data pada PostgreSQL.
